## GDP Types Visualization

```python
def visualize_GDP(input_gdp, title, country_codes=[ 'TUR', 'POL', 'GRC'], year_intervals=['1995', '2019'], unit='Dollars'):
    
    title = f'{title} during {year_intervals[0]} - {year_intervals[1]}'
    
    first_year = int(year_intervals[0])
    last_year = int(year_intervals[1])
    
    
    plt.figure(figsize=(24,6))
    plt.title(label=title,fontsize=20,fontweight=20)
    plt.ylabel(unit,fontsize=25)
    plt.xlabel('Years',fontsize=25)
    for code in country_codes:
        plt.plot(np.arange(first_year,last_year),input_gdp.loc[input_gdp['Code'] == code].iloc[0,37:-2],linewidth=3,label=code)

    plt.xticks(np.arange(first_year,last_year),np.arange(first_year,last_year),rotation=0)
    plt.grid()
    plt.legend()
    plt.show()
    print('\n')
```

```python
visualize_GDP(gdp, 'GDP')
visualize_GDP(gdp_per_capita, 'GDP per capita')
visualize_GDP(gdp_ppp, 'GDP PPP')
visualize_GDP(gdp_ppp_per_capita, 'GDP PPP per capita')

visualize_GDP(gdp_growth, 'GDP Growth(Annually)', unit='Ratio')
visualize_GDP(gdp_per_capita_growth, 'GDP per capita Growth(Annually)', unit='Ratio')
```

## Calculate GDP growth between years in %

```python
def calculate_gdp_growth_between_years(input_df, start_year, end_year, country_code):
    first_gdp = float(input_df[input_df['Code']==country_code][start_year])
    last_gdp = float(input_df[input_df['Code']==country_code][end_year])
    growth = ((last_gdp - first_gdp) / (last_gdp + first_gdp)) * 100
    return round(growth, 4)
```

```python
calculate_gdp_growth_between_years(gdp, '2001', '2013', 'KOR')
```
